```python
import re

def remove_tags(raw_text):
  cleaned_text = re.sub(re.compile('<.*?>'), '', raw_text)
  return cleaned_text

df.loc[:, 'review'] = df['review'].apply(remove_tags)
```

```python
df
```

```python
df['review']=df['review'].apply(lambda x:x.lower())
#
```

```python
import nltk
from nltk.stem.snowball import stopwords
nltk.download('stopwords')

sw_list=stopwords.words('english')
df.loc[:, 'review'] = df['review'].apply(lambda x: [item for item in x.split() if item not in sw_list]).apply(lambda x:" ".join(x))
```

```python
df
```

```python
x=df.iloc[:,0:1]
y=df['sentiment']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
y=encoder.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
x_train_bow=cv.fit_transform(x_train['review'].values)
x_test_bow=cv.transform(x_test['review'].values)
```

```python
x_train_bow.shape
```

```python
x_test_bow.shape
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()
mnb.fit(x_train_bow, y_train)
```

```python
y_pred=mnb.predict(x_test_bow)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
#
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
rf.fit(x_train_bow,y_train)
```

```python
rf.fit(x_train_bow,y_train)
y_pred=rf.predict(x_test_bow)
accuracy_score(y_test,y_pred)
```

```python
cv=CountVectorizer(max_features=3000)
x_train_bow=cv.fit_transform(x_train['review'].values)
x_test_bow=cv.transform(x_test['review'].values)
rf=RandomForestClassifier()
rf.fit(x_train_bow,y_train)
Y_PRED=rf.predict(x_test_bow)
accuracy_score(y_test,Y_PRED)
```

```python
cv=CountVectorizer(max_features=3000,ngram_range=(1,2))

x_train_bow=cv.fit_transform(x_train['review'].values)
x_test_bow=cv.transform(x_test['review'].values)

rf=RandomForestClassifier()
rf.fit(x_train_bow,y_train)
y_pred=rf.predict(x_test_bow)
accuracy_score(y_test,y_pred)
```
