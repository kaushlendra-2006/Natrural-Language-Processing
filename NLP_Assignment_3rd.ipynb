**Assignment 3:** Apply NER to identify entities such as names, organizations, and locations in a given text. Perform sentence segmentation on a paragraph and explain its importance in NLP tasks.

```python
import spacy

# Load English NLP model
nlp = spacy.load("en_core_web_sm")

```

```python
# Input text
text = """Artificial Intelligence has become one of the most influential technologies in the modern world. In 2023, Sundar Pichai, the CEO of Google, announced several AI-driven initiatives during a global technology conference held in San Francisco. These initiatives aim to improve healthcare, education, and transportation systems across the United States.

Meanwhile, Microsoft, headquartered in Redmond, has been investing heavily in cloud computing and artificial intelligence through its collaboration with OpenAI. Researchers from Stanford University and Massachusetts Institute of Technology have also contributed significantly to AI research.

In India, cities like Bengaluru and Hyderabad have emerged as major technology hubs, attracting global companies such as Amazon, IBM, and Infosys. Governments and organizations around the world continue to explore ethical and responsible uses of artificial intelligence to ensure sustainable development for future generations."""
```

```python
# Process text
doc = nlp(text)
```

```python
# -------- Sentence Segmentation --------
print("Sentence Segmentation:")
for sent in doc.sents:
    print("-", sent.text)
```

```python
# -------- Named Entity Recognition --------
print("\nNamed Entity Recognition:")
for ent in doc.ents:
    print(f"Entity: {ent.text} | Label: {ent.label_}")
```
