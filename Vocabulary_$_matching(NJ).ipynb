Rule -based Matching

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
from spacy.matcher import Matcher
matcher=Matcher(nlp.vocab)
```

```python
pattern1=[{'LOWER':'solarpower'}]
pattern2=[{'LOWER':'solar'},{'LOWER':'power'}]
pattern3=[{'LOWER':'solar'},{'IS_PUNCT':True},{'LOWER':'power'}]

matcher.add('SolarPower',[pattern1,pattern2,pattern3])
```

```python
# Create a sample document
doc = nlp("The Solar power industry continues to grow as demand for solar power increases.")

# Apply the matcher to the document
matches = matcher(doc)

# Print the matches
for match_id, start, end in matches:
    string_id = nlp.vocab.strings[match_id]  # Get string representation
    span = doc[start:end]  # Get the matched span
    print(match_id, string_id, start, end, span.text)
```

```python
pattern1 = [{'LOWER': 'solarpower'}]
pattern2 = [{'LOWER': 'solar'}, {'IS_PUNCT': True, 'OP': '*'}, {'LOWER': 'power'}]

matcher.add('SolarPower', [pattern1, pattern2])
```

```python
doc2 = nlp(u'Solar--power is solarpower energy runs solar-powered cars.')
```

```python
found_matches = matcher(doc2)
print(found_matches)
```

PhraseMatcher

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
```

```python
# Create the file and write some content
with open('reaganomics.txt', 'w') as f:
    f.write('This is a sample text about reaganomics.')

# Now, open and read the file
with open('reaganomics.txt','r') as f:
    doc3 = nlp(f.read())
```

```python
import os
print(os.listdir())
```

```python
# first,create a list of match phrase:
phrase_list = ['voodoo economics','supply-side economics','trickle-down economics','free-market economics']

#next, convert each phrase to a Doc object :
phrase_patterns = [nlp(text) for text in phrase_list]

#pass each Doc object into matcher (mote the use of the asterisk!):
matcher.add('VoodooEconomics', None, *phrase_patterns)

#build a list of matches:
matches = matcher(doc3)

print(matches)
```
