Subword Tokenization

```python
string="I am learning Tokenization"
```

```python
newString=string.split()
```

```python
newString
```

```python
List3=[]
for word in newString:
  if word[-3:]=="ing":
    List3.append(word.partition("ing"))
  elif word[-7:]=="ization":
    List3.append(word.partition("ization"))
  else:
    List3.append(word)
print(List3)
```

```python
list4=[]
for element in List3:
  if type(element) is tuple:
    list4.append(element[0])
    list4.append(element[1])
  else:
    list4.append(element) # Append the element directly if it's not a tuple
print(list4)
```

Visualizing the dependary parse

```python
from spacy import displacy

doc=nlp(u'Apple is going to build a U.K. factory for $6 million.')
displacy.render(doc,style='dep',jupyter=True,options={'distance':110})
```

Visualizing the entity recognizer

```python
doc = nlp("over the last quarter Apple sold nearly  20 thousand  iPods for a profit of $6 million.")
displacy.render(doc,style='ent',jupyter=True)
```
